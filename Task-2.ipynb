```python
#EDA STEP 1: KNOWING INITAL INFORMATION ABOUT DATA
#STEP1- knowing initial details about data set
#import the libraries
import pandas as pd 
import seaborn as sns 
#importing car dataset downloaded from Kaggle
#here data frame by name car is created
cars=pd.read_csv("C:\\Users\\taran\\Data Science\\PRODIGY-INFOTECH-TARAN PANFAIR\\automobile.csv")
#shows how many rows and columns are there in data set
print(cars.shape)
#cars.info we get dataset information
cars.info()
print(cars.columns)
#to display top 5 rows
cars.head()
#to display n rows
cars.head(7)
#to display last 5 rows
cars.tail()
#to display n rows
cars.tail(10)
print(list(cars.columns))#to display in list

```

```python
#EDA STEP 2: MODIFYING OR REMOVING UNWANTED DATA
#remove duplicate rows if ny
#keep first row and remove other duplicate rows of that row
cars=cars.drop(['cylinders','displacement'],axis=1)
print(cars.shape)
print(cars.columns)
#drop duplicate command
#keep first row and remove duplicate of that row
c=cars.drop_duplicates(keep='first')
print(c)
#finding total no of rows which are missing value in a column
#isnull command
a=cars.isnull().sum()
print(a)
#sorting data wrt a column-- here we sort mpg on ascending order
cars_sort=cars.sort_values(by='mpg', ascending=True)
print(cars_sort)
```

```python
##EDA STEP 3: RETREIVING DATA
#select dt_types command
#display columns only having float values
cars1=cars.select_dtypes(include=['float64'])
print(cars1)
#describe command
#describe command is used for obtaining some statistical data like count,mean etc
cars.describe()
#iloc[] command- gives integer location based indexing or selection
#display 0th row
c=cars.iloc[0]
print(c)
#display last row
c=cars.iloc[-1]
#select row from 0th to 4th
c=cars.iloc[0:5]
#select all rows and 0th column
c=cars.iloc[:,0]
#loc[] command-gives integer location based indexing or selection
#display row 0 to row 5 with column name mpg
c1=cars.loc[0:5,'mpg']
print(c1)
```
